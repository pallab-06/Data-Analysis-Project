# EDA

## 1. Popular Internship Titles

```python
# Count occurrences of each unique internship title
title_counts = df['internship_title'].value_counts()

# Display the most popular internship titles
print(title_counts.head(10))
```

```python
# Plot the most popular internship titles
plt.figure(figsize=(10, 6))
title_counts.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Popular Internship Titles')
plt.xlabel('Internship Title')
plt.ylabel('Number of Occurrences')
plt.show()
```

## 2. Company Engagement

```python
# Count occurrences of each unique company name
company_counts = df['company_name'].value_counts()

print(company_counts.head(10))
# Plot the companies offering the most internships
plt.figure(figsize=(10, 6))
company_counts.head(10).plot(kind='bar', color='green')
plt.title('Top 10 Companies Offering the Most Internships')
plt.xlabel('Company Name')
plt.ylabel('Number of Internships Offered')
plt.show()
```

## 3. Geographic Distribution:

```python
# Count occurrences of each unique location
location_counts = df['location'].value_counts()


# Plot the geographic distribution of internships
plt.figure(figsize=(12, 8))
location_counts.head(10).plot(kind='bar', color='orange')
plt.title('Geographic Distribution of Internships')
plt.xlabel('Location')
plt.ylabel('Number of Internships')
plt.show()
```

## 4. Start Date Distribution

```python
# Count occurrences of each unique start date value
startdate_counts = df['start_date'].value_counts()

# Plot the distribution of start date values
plt.figure(figsize=(8, 6))
startdate_counts.plot(kind='bar', color='blue')
plt.title('Distribution of Start Date Values')
plt.xlabel('Start Date')
plt.ylabel('Number of Occurrences')
plt.show()
```

## 5. Internship Duration Distribution

```python
# Plot the distribution of internship durations
plt.figure(figsize=(10, 6))
plt.hist(df['duration'], bins=20, color='green', edgecolor='black')
plt.title('Distribution of Internship Durations')
plt.xlabel('Duration (Days or Weeks)')
plt.ylabel('Number of Internships')
plt.grid(True)
plt.show()
```

## 6. Stipend Distribution

```python
df['stipend'].unique()
```

```python
# Replace 'Unpaid' with 0 in the stipend column
df['stipend'] = df['stipend'].apply(lambda x: '0' if 'Unpaid' in str(x) else x)

# Remove unwanted substrings and symbols
df['stipend'] = df['stipend'].str.replace("/month", "").str.replace("lump sum", "").str.replace("₹", "")

# Extract other benefits and split stipend into min and max columns
df['Other Benifits'] = df['stipend'].str.split('+').apply(lambda x: x[-1].strip() if len(x) > 1 else None)
df['stipend'] = df['stipend'].str.split('+').str[0].str.strip()

# Extract numeric values and handle commas
df['min stipend'] = df['stipend'].str.split("-").str[0].str.extract('([\d,]+)', expand=False).str.replace(",", "")
df['max stipend'] = df['stipend'].str.split("-").str[-1].str.extract('([\d,]+)', expand=False).str.replace(",", "")

# Convert numeric columns to numeric type
numeric_columns = ['min stipend', 'max stipend']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
df['stipend'].unique()
```

```python
# Plot the distribution of minimum stipend values
plt.figure(figsize=(12, 6))
sns.histplot(df['min stipend'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Minimum Stipend Values')
plt.xlabel('Minimum Stipend (in INR)')
plt.ylabel('Frequency')
plt.show()

# Plot the distribution of maximum stipend values
plt.figure(figsize=(12, 6))
sns.histplot(df['max stipend'], bins=30, kde=True, color='salmon')
plt.title('Distribution of Maximum Stipend Values')
plt.xlabel('Maximum Stipend (in INR)')
plt.ylabel('Frequency')
plt.show()

```

## 7. Investigating if there's a correlation between the stipend amount and the duration of internships.

```python
df['duration'].unique()
```

```python
# Conversion factors
days_in_month = 30  # Assuming an average of 30 days in a month
days_in_week = 7

# Convert duration to numeric values in days
df['duration_numeric'] = df['duration'].apply(lambda x: int(x.split()[0]) * days_in_month if 'Month' in x else int(x.split()[0]) * days_in_week)
```

```python
# Scatter plot to visualize the relationship between stipend and duration
plt.figure(figsize=(12, 6))
sns.scatterplot(x='min stipend', y='duration_numeric', data=df, hue='max stipend', palette='viridis', size='max stipend')
plt.title('Correlation between Stipend and Duration of Internships')
plt.xlabel('Minimum Stipend (in INR)')
plt.ylabel('Duration (in days)')
plt.legend(title='Maximum Stipend')
plt.show()

# Correlation matrix to quantify the correlation between stipend and duration
correlation_matrix = df[['min stipend', 'max stipend', 'duration_numeric']].corr()

# Heatmap for better visualization of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix: Stipend and Duration')
plt.show()
```

## 8. Companies that offer unpaid internships

```python
# Identify companies that offer unpaid internships
unpaid_companies = df[df['min stipend'] == 0]['company_name'].unique()

# Print or use the unpaid_companies as needed
print("Companies offering unpaid internships:")
print(unpaid_companies)
```

## 9. Analyzing how internship titles are distributed within each company to identify patterns.

```python
# Group by Company and Title
title_distribution = df.groupby(['company_name', 'internship_title']).size().reset_index(name='title_count')

# Explore Top Internship Titles
top_titles = title_distribution.sort_values(by='title_count', ascending=False).groupby('company_name').head(1)

# Explore Unique Title Counts
unique_title_counts = title_distribution.groupby('company_name')['internship_title'].nunique()

# Visualize Top 10 Internship Titles
top_10_titles = title_distribution.sort_values(by='title_count', ascending=False).head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x='title_count', y='internship_title', data=top_10_titles, palette='viridis')
plt.title('Top 10 Internship Titles within Companies')
plt.xlabel('Title Count')
plt.ylabel('Internship Title')
plt.show()

# Print or use the results as needed
print("Top Internship Titles:")
print(top_titles)

print("\nUnique Title Counts within Companies:")
print(unique_title_counts)
```

## 10. Top-Paying Companies

```python
# Calculate average stipend for each company
average_stipend = df.groupby('company_name')['min stipend'].mean().reset_index(name='average_stipend')

# Identify top-paying companies
top_paying_companies = average_stipend.sort_values(by='average_stipend', ascending=False).head(10)
print(top_paying_companies)

# Plot the top 10 paying companies
plt.figure(figsize=(12, 8))
sns.barplot(x='average_stipend', y='company_name', data=top_paying_companies, palette='viridis')
plt.title('Top 10 Paying Companies')
plt.xlabel('Average Stipend Amount')
plt.ylabel('Company Name')
plt.show()
```
